```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
pd.options.display.max_rows = 999
```

```python
np.random.seed(0)
```

```python
data = load_iris()
```

```python
data
```

```python
dat = pd.DataFrame(data.data , columns= data.feature_names)
```

```python
dat.head()
```

```python
dat['species'] = pd.Categorical.from_codes(data.target,data.target_names)
# it is for fetching the data from the targets and then fetching the data from the targeted names for a given target 
```

```python
dat.head()
```

```python
dat["is_train"] = np.random.uniform(0,1, len(dat)) <= 0.75 
# here we choose random n0. between 0 and 1 and classify them as true if less than 0.75 and false if not 
dat
```

```python
train ,test = dat[dat['is_train']==True] , dat[dat['is_train'] == False] # train test split
```

```python
len(train) # these are not same as shown in the video becz of the randomness
```

```python
train
```

```python
len(test)
```

```python
features = dat.columns[0:4]
features
```

```python
y = pd.factorize(train['species'])[0]
```

```python
y
```

```python
clf = RandomForestClassifier(n_jobs= 2 , random_state=0)
clf.fit(train[features],y)
```

```python
clf.predict(test[features])
```
